# kmeans

```python
#-*- coding: utf-8 -*-
#K-Means聚类算法
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
from sklearn.cluster import KMeans #导入K均值聚类算法

inputfile = '201404.xlsx' #待聚类的数据文件
k = 50                       #需要进行的聚类类别数

#读取数据并进行聚类分析
data = pd.read_excel(inputfile) #读取数据

#调用k-means算法，进行聚类分析
kmodel = KMeans(n_clusters = k, n_jobs = 2) #n_jobs是并行数，一般等于CPU数较好
kmodel.fit(data) #训练模型
```

```python
data['label']=kmodel.labels_
```

```python
kcenter=pd.DataFrame(kmodel.cluster_centers_)
```

```python
figure=plt.figure(figsize=(12,8))
ax1=figure.add_subplot(1,1,1)
ax1.scatter(kcenter[0],kcenter[1],c='k',marker='v',label='center')
ax1.scatter(data['X2'],data['Y2'],c=data['label'],alpha=0.3,label='point')
ax1.set_title("kmeans")
ax1.legend(loc='best')
```
